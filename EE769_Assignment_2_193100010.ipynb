```python
# EE769 Assignment 2


#import numpy library
import numpy as np # 
# import pandas library
import pandas as pd 

# Importing classifiers

from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import ExtraTreesClassifier


# Get training data from folder where it is saved
training = pd.read_csv("/tmp/mozilla/train.csv")
training.head()

#Get testing data from folder where it is saved
testing = pd.read_csv("/tmp/mozilla/test.csv")
testing.head()

# Grouping age values to 4 groups. Value within certian class will attain certian value.

def Age(dataframe):
    dataframe.loc[dataframe['Age'] <= 30,'Age'] = 1
    
    dataframe.loc[dataframe['Age'] <= 40 & (dataframe['Age'] > 30) , 'Age'] = 2
    
    dataframe.loc[dataframe['Age'] <= 50 & (dataframe['Age'] > 40) , 'Age'] = 3
    
    dataframe.loc[dataframe['Age'] <= 60 & (dataframe['Age'] > 50) , 'Age'] = 4
    
    return dataframe


# All data not in numerical form. BusinessTravel, Department, EducationField, Gender, JobRole, MaritalStatus, OverTime are in string or categorical form
# Using Scikit-Learn label encoding is done for this categorical data. A number assigned to each entry as encoded by encoder.

from sklearn.preprocessing import LabelEncoder
encode = LabelEncoder()
training['BusinessTravel'] = encode.fit_transform(training['BusinessTravel'])
testing['BusinessTravel'] = encode.fit_transform(testing['BusinessTravel'])

training['Department'] = encode.fit_transform(training['Department'])
testing['Department'] = encode.fit_transform(testing['Department'])

training['EducationField'] = encode.fit_transform(training['EducationField'])
testing['EducationField'] = encode.fit_transform(testing['EducationField'])

training['Gender'] = encode.fit_transform(training['Gender'])
testing['Gender'] = encode.fit_transform(testing['Gender'])

training['JobRole'] = encode.fit_transform(training['JobRole'])
testing['JobRole'] = encode.fit_transform(testing['JobRole'])

training['MaritalStatus'] = encode.fit_transform(training['MaritalStatus'])
testing['MaritalStatus'] = encode.fit_transform(testing['MaritalStatus'])

training['OverTime'] = encode.fit_transform(training['OverTime'])
testing['OverTime'] = encode.fit_transform(testing['OverTime'])

# Calling age function for both training and testing dataset

Age(training); 
Age(testing);


#Definign attrition/output of training data 

ytraining = training["Attrition"]

# Defining features  without output 'Age,BusinessTravel,DailyRate,Department,DistanceFromHome,Education,EducationField,EmployeeCount,,EnvironmentSatisfaction,Gender,HourlyRate,JobInvolvement,JobLevel,JobRole,JobSatisfaction,MaritalStatus,MonthlyIncome,MonthlyRate,NumCompaniesWorked,OverTime,PercentSalaryHike,PerformanceRating,RelationshipSatisfaction,StockOptionLevel,TotalWorkingYears,TrainingTimesLastYear,WorkLifeBalance,YearsAtCompany,YearsInCurrentRole,YearsSinceLastPromotion,YearsWithCurrManager
# exclusing features EmployeeNumber, ID aas they not affect attrition.

features = ['Age', 'BusinessTravel', 'DailyRate', 'Department','DistanceFromHome', 'Education', 'EducationField', 'EnvironmentSatisfaction', 'Gender', 'HourlyRate','JobInvolvement', 'JobLevel', 'JobRole', 'JobSatisfaction',     'MaritalStatus', 'MonthlyIncome', 'MonthlyRate', 'NumCompaniesWorked', 'OverTime', 'PercentSalaryHike', 'PerformanceRating', 'RelationshipSatisfaction', 'StockOptionLevel', 'TotalWorkingYears','TrainingTimesLastYear', 'WorkLifeBalance', 'YearsAtCompany','YearsInCurrentRole', 'YearsSinceLastPromotion', 'YearsWithCurrManager']

# Defining X values or inputs values of training and testing data. It is without attrition and contains features defined above
Xtraining = training[features]
Xtesting = testing[features]

#Using Extra Trees Classifier  

clf = ExtraTreesClassifier(n_estimators=200, max_depth=None ,random_state=0)
clf.fit(Xtraining, ytraining)
# Determining predictions
predictions = clf.predict(Xtesting)

# taking output frpm data frame 

output = pd.DataFrame({'ID': testing.ID, 'Attrition': predictions})

# Saving output csv file to local address

output.to_csv('/tmp/mozilla/EE769_193100010_ETC_final2.csv', index=False)

# File named 'EE769_193100010_ETC_final2.csv' stored at local address defined above
```
